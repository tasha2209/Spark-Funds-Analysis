```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb
import scipy.stats as st
```

```python
wiki=pd.read_html(r'https://en.wikipedia.org/wiki/List_of_territorial_entities_where_English_is_an_official_language')
```

```python
country1=wiki[0]
country1=country1.loc[0:]
```

```python
country2=wiki[1]
```

```python
country1=country1.loc[:,['Alpha-3 code','Primary language?']]
country1
```

```python
country2=country2.loc[:,['Alpha-3 code','Primary language?']]
country2
```

```python
master=pd.concat([country1,country2])
```

```python
master.reset_index(drop='index',inplace=True)
```

```python
master
```

```python
master.rename(columns={'Alpha-3 code':'country_code','Primary language?':'Official_lang'},inplace=True)
```

```python
master['Official_lang'] = master['Official_lang'].apply(lambda x:x.split()[0])
```

```python
master
```
